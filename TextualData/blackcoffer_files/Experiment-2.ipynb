```python
def get_text_bs(html):
    tree = BeautifulSoup(html, 'lxml')

    body = tree.body
    if body is None:
        return None

    for tag in body.select('script'):
        tag.decompose()
    for tag in body.select('style'):
        tag.decompose()

    text = body.get_text(separator='\n')
    return text
```

```python
response = urllib.request.urlopen('https://www.sec.gov/Archives/edgar/data/12239/0001104659-07-062470.txt')
content = response.read().decode('utf8')
```

```python
new_content = get_text_bs(content)
```

```python
re.sub(r'\n|\t|(\xa0)|\(x\)\d+', " ", new_content)
```

```python
import requests
from bs4 import BeautifulSoup

url = ''
res = requests.get(url)
html_page = res.content
soup = BeautifulSoup(html_page, 'html.parser')
text = soup.find_all(text=True)

output = ''
blacklist = [
    '[document]',
    'noscript',
    'header',
    'html',
    'meta',
    'head', 
    'input',
    'script',
    's'
    # there may be more elements you don't want, such as "style", etc.
]

for t in text:
	if t.parent.name not in blacklist:
		output += '{} '.format(t)

print(output)
```

```python
output = output.lower()
```

```python
output = re.sub(r'\n|\t|\xa0', ' ', output)
```

```python
output = re.sub(r"[^a-zA-Z\s\.]", "", output)
```

```python
output = re.sub(r"\s+", " ", output)
```
